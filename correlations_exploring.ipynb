```python
import pandas as pd
df = pd.read_csv('cleaned_data.csv')
df.head()
```

```python
df = df.drop(['participant_id', 'day'], axis=1)
```

```python
pss_correlations = df.corr()['PSS_score'].drop('PSS_score')  # Exclude self-correlation

print("Correlation of PSS with other attributes:")
print(pss_correlations)

pss_correlations = pss_correlations.sort_values(ascending=False)
print("\nSorted Correlations:")
print(pss_correlations)
```
