```python
import pandas as pd

df = pd.read_csv('stress_detection.csv')
df.head(50)
```

```python
#Data cleaning procedure
#1. Analyzing columns independently
```

```python
#a. Identyfing the missing value

df.isnull().sum() # no missing value are detected
```

```python
#b. Fix Inconsistent or Logical Errors
# b.1. column sleep time seems too have non-logical values, new values for sleep_time: at which time a person went to sleep
df['sleep_time'] = df['wake_time'] + df['sleep_duration']
df['sleep_time'] = df['sleep_time'].apply(lambda x: 12-x if x < 12 else 24-(x-12))
```

```python
df.head()
```

```python
#c. check the range of the columns that have it

out_of_range_day = df[~df['day'].between(1, 30)]
print(out_of_range_day)

out_of_range_PSS = df[~df['PSS_score'].between(0, 40)]
print(out_of_range_PSS)

out_of_range_sleep_time = df[~df['sleep_time'].between(0, 24)]
print(out_of_range_sleep_time)

out_of_range_PSQI= df[~df['PSQI_score'].between(0, 21)]
print(out_of_range_PSQI)
```

```python
# Normalize data ranges for continuous variables like skin_conductance or accelerometer to bring values into comparable scales:

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[['skin_conductance', 'accelerometer']] = scaler.fit_transform(df[['skin_conductance', 'accelerometer']])
```

```python
#e. check for duplicates
df[df.duplicated()]
```

```python
Q1 = df['screen_on_time'].quantile(0.25)
Q3 = df['screen_on_time'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['screen_on_time'] < lower_bound) | (df['screen_on_time'] > upper_bound)]
outliers
```
